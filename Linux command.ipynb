2.) What is the difference between Hard Link & Soft LinkE


Answer :-  A hard link is a direct reference to the same data (inode) as the original file. When you create a hard link, you are creating another file name that points to the exact same content. Both files share the same inode number. Deleting the original file does not delete the data, because the hard link still points to it. Hard links only work within the same filesystem and cannot be created for directories (by default).

3.) What is a Kernel in LinuxE ?

ANS:-  The kernel is the core part of the Linux operating system., It acts as a bridge between hardware and software.
 Key Functions of the Kernel:

 -Memory management: Keep track of how much memory is used to store what, and where
 - Process management: Determine which processes can use the central processing unit (CPU), when, and for how long
 - Device drivers: Act as mediator/interpreter between the hardware and processes
 - System calls and security: Receive requests for service from the processes

 we can check the kernel name, version, and other system information using the uname command in Linux.


*   uname – This shows the kernel name (usually Linux).

*   uname -r – This shows the kernel version, which tells you which version of the Linux kernel is currently running.

*    uname -v – This gives the kernel build info.

*   uname -m – Shows the machine architecture, like x86_64.


*   uname -a – This displays all system info in one line: kernel name, version, build, architecture, and more.

5.)  What is the ‘grep’ command used for in LinuxE

ANS->  The grep command is used to search for specific patterns, words, or text within files or output in Linux.

It stands for “Global Regular Expression Print.”

| Example Command              | What It Does                                      |                                                  |
| ---------------------------- | ------------------------------------------------- | ------------------------------------------------ |
| `grep "error" logfile.txt`   | Finds lines containing `"error"` in `logfile.txt` |                                                  |
| \`ps aux                     | grep firefox\`                                    | Filters running processes to show only `firefox` |
| `grep -i "hello" myfile.txt` | Case-insensitive search for `"hello"`             |                                                  |
| `grep -r "main()" .`         | Recursively searches for `"main()"` in all files  |                                                  |
| `grep -v "skip"` file.txt    | Shows all lines **not** containing `"skip"`       |                                                  |
